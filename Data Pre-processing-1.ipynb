### Here columns 'WiFi' and 'Blutooth' does not have desired data for most of the rows, so we will drop them

```python
df = df.drop(['Wifi', 'Blutooth'], axis=1)
```

## Coverte The Catagorical Data into Numaric

### ROM

#### let's extract only usable part of the string data & convert it into int

```python
df['RAM'] = df['RAM'].str.replace(' GB', '').astype(int)
```

### Brand

let's use LabelEncoder for conversion

```python
label_encoder = LabelEncoder()

df['Brand'] = label_encoder.fit_transform(df['Brand'])
```

### Rear Camera

Craete a function which will take Rear Camera values as input and return the Total Rear Camera Resolution in int

```python
def get_total_rear_camera_resolution(str):
    return sum([int(num) for num in re.findall(r'\d+', str)])
 
df['Rear Camera'] = np.array([get_total_rear_camera_resolution(i) for i in df['Rear Camera'].values]) 
```

### Front Camera

#### let's extract only usable part of the string data & convert it into float

```python
df['Front Camera'] = df['Front Camera'].str.split().str[0].astype(float)
```

### Battery

#### let's extract only usable part of the string data & convert it into float

```python
df['Battery'] = df['Battery'].str.split().str[0].astype(float)
```

### Display Size

```python
df['Display Size'] = df['Display Size'].str.split().str[0].astype(float)
```

### Resolution

```python
def get_total_display_resolution(str):
    return sum([int(num) for num in re.findall(r'\d+', str)])

df['Resolution'] = np.array([get_total_display_resolution(i) for i in df['Resolution'].values]) 
```

### Screen To Body Ratio

```python
df['Screen To Body Ratoio'] = df['Screen To Body Ratoio'].str.split().str[0].astype(float)
```

### ROM

```python
df['ROM'] = df['ROM'].str.split().str[0].astype(int)
```

### Sim Slot

```python
df['Sim Slot'] = label_encoder.fit_transform(df['Sim Slot'].str.split().str[0])
```

### FingerPrint Sensor

```python
df['FingerPrint Sensor'] = label_encoder.fit_transform(df['FingerPrint Sensor'].str.split().str[0])
```

### Price

```python
df['Price'] = df['Price'].str.replace('Rs.', '').str.replace(',', '').astype(float) 
```

### Processor

```python
def calculate_total_power(st):
    words = st.split()
    unneeded = ['core', 'Core', 'core)', 'Core)', 'Core,', 'core,', 'GHz', 'GHz,', '+']
    filtered_words = [word for word in words if word not in unneeded]
    arr = pd.Series([1, 2, 3, 4, 6, 8], index=['Single', 'Dual', 'Tri', 'Quad', 'Hexa', 'Octa']) 
    powers = [arr[word] for word in filtered_words if word in arr.index ]
    temp = [word for word in filtered_words if word not in arr.index]
    values = [float(word.split('(')[-1]) for word in temp]
    
    if(len(temp) != 1):
        powers.remove(powers[0])
    
    return sum(a * b for a, b in zip(powers, values))

df['Processor'] = np.array([calculate_total_power(i) for i in df['Processor'].values])
```
