# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read Data

```python
df = pd.read_csv('Mobile-DataSet-Pre-Processed.csv')
df.sample(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.dtypes
```

```python
plt.hist(df['Price'],bins=15, rwidth=0.90)
plt.xlabel('Price')
plt.ylabel("Number Of Phones")
```

```python
plt.scatter(df['Front Camera'], df['Price'])
plt.xlabel('Front Camera ')
plt.ylabel("Price")
plt.plot() 
```

```python
plt.boxplot(df['Front Camera'])
plt.ylabel('MP')
plt.xlabel('Front Camera')
```

```python
plt.scatter(df['Display Size'], df['Price'], c='orange')
plt.xlabel('Display Size')
plt.ylabel("Price")
plt.plot()
```
