## Correlation between 'Front Camera' and 'Price'

```python
plt.scatter(df['Front Camera'], df['Price'])
plt.xlabel('Front Camera ')
plt.ylabel("Price")
plt.plot() 
```

```python
df['Front Camera'].describe()
```

#### as we can see above, there is an outlier

### let's Remove Outliers

```python
Q1 = np.percentile(df['Front Camera'], 25)
Q3 = np.percentile(df['Front Camera'], 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
  
df = df[(df['Front Camera'] >= lower_bound) & (df['Front Camera'] <= upper_bound)] 
df.info()
```

```python
plt.scatter(df['Front Camera'], df['Price'], c='pink')
plt.xlabel('Front Camera ')
plt.ylabel("Price")
plt.plot() 
```
